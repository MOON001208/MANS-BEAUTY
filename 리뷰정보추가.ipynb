```python
# 1. 가격 매핑 사전 정의 (상품명 키워드 : 변경할 가격)
price_map = {
    '다슈 맨즈 아쿠아 매트 비비 쿠션 15g': 21000,
    '다슈 맨즈 아쿠아 매트 비비쿠션 (본품+리필)': 30000,
    '다슈 맨즈 아쿠아 매트 비비 쿠션 (리필용)': 18000,
    '그라펜 팁 스케치 멀티 커버 파운데이션': 32000,
    '듀이셀 트루빔 파운데이션': 36000,
    '라끌랑 슈퍼쉴드 옴므쿠션': 39000,
    '라끌랑 올데이 파워 픽스쿠션': 37000,
    '비레디 블루 파운데이션': 29000,
    '오브제 내추럴 커버 파운데이션': 36800,
    '원오브뎀 드 뗑 쿠션': 32000,
    '정샘물 맨 슬림핏 누더 쿠션': 34000
}

# 2. 가격 업데이트 함수 정의
def update_price(row):
    product_name = row['상품이름']
    for key, price in price_map.items():
        # 상품이름에 키워드가 포함되어 있는지 확인
        if key in product_name:
            return price
    return row['가격']  # 매칭되는 게 없으면 기존 가격 유지

# 3. 데이터프레임에 적용
df['가격'] = df.apply(update_price, axis=1)

# 결과 확인
print(df[['브랜드', '상품이름', '가격']])
```

```python
import pandas as pd
import re

def clean_options(option_list_str):
    if pd.isna(option_list_str) or option_list_str == "[None]":
        return None

    # 1. 텍스트 정돈 (리스트 형태의 문자열일 경우 대비)
    # [옵션], 기획, 증정, 리필 등의 노이즈 단어 제거
    text = str(option_list_str)
    text = re.sub(r'\[옵션\]|기획|증정|리필|본품|None|[\'\"\[\]|맨퍼프2P증정]', '', text)
    text = str(option_list_str).replace("[", "").replace("]", "").replace("'", "")
    # 2. 호수 패턴 추출 (예: 1호, 01, 21호, 21, 001 등)
    # 숫자 뒤에 '호'가 붙거나, 독립적인 숫자(호수)를 찾음
    finds = re.findall(r'(\d+호|\d+)', text)

    # 3. 공백 제거 및 중복 제거
    # '호'를 유지하기 위해 int 변환 없이 strip()만 수행   
    cleaned = [f.strip() for f in finds]
    
    # 4. 중복 제거 (set 사용) 후 정렬
    return ", ".join(cleaned) if cleaned else None

# --- 데이터프레임 적용 ---

# 예시: '옵션' 컬럼에 위에서 보여주신 리스트들이 들어있다고 가정
# df['호수_리스트'] = df['옵션'].apply(clean_options)
# df['호수_개수'] = df['호수_리스트'].apply(len)
```

```python
df['호수'] = df['옵션'].apply(clean_options)
```

```python
import pandas as pd
import re

def clean_options_smart(option_list_str):
    # 1. 입력값이 없거나 [None]인 경우 처리
    if pd.isna(option_list_str) or str(option_list_str) == "[None]":
        return None

    # 2. 문자열을 개별 옵션으로 분리
    # 제공해주신 데이터 형태(['옵션1' '옵션2'])를 고려하여 텍스트 정돈
    text = str(option_list_str)
    text = re.sub(r'\[옵션\]|기획|증정|리필|본품|None|[\'\"\[\]|맨퍼프2P증정]', '', text)
    text = str(option_list_str).replace("[", "").replace("]", "").replace("'", "")
    options = text.split("' '") # 또는 공백이나 콤마 기준으로 분리 시도
    if len(options) == 1: # 분리가 안 된 경우 공백으로 재시도
        options = re.split(r' (?=\[옵션\])', text)

    cleaned_res = []
    
    for opt in options:
        # [옵션] 글자 뒤부터 검사하기 위해 텍스트 정리
        opt = opt.replace("[옵션]", "").strip()
        
        # 3. 각 옵션 문구에서 "가장 처음 나타나는" 숫자+호 또는 숫자 찾기
        # \d+호 : 1호, 21호 등
        # \d+   : 01, 02 등 (호가 없는 경우)
        match = re.search(r'(\d+호|\d+)', opt)
        
        if match:
            # 첫 번째로 매칭된 결과만 가져오고 반복 중단 (15g 같은 뒤쪽 숫자 방지)
            found = match.group(1)
            cleaned_res.append(found)

    # 4. 중복 제거 및 정렬
    return ", ".join(cleaned_res) if cleaned_res else None

```

```python
df['호수'] = df['옵션'].apply(clean_options_smart)
```
