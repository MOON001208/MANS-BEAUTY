23937개 데이터에서 17558개가 되었음 딱 6379개 제거

duplicated로 정제해도 완전히 똑같지 않으면 제거가 되지 않아서 자카드 유사도 이용해서 제거

보통 1달 후기 리뷰같은 경우는 기존 리뷰에서 단어 조금만 바꿔서 올리기 때문에 이와같은 과정을 거쳐야함

```python
import pandas as pd
from itertools import combinations

# 자카드 유사도 함수 (공백 무시)
def get_jaccard_sim(str1, str2):
    if not str1 or not str2: return 0
    a = set(str1.replace(" ", ""))
    b = set(str2.replace(" ", ""))
    return len(a & b) / len(a | b) if len(a | b) != 0 else 0

# 1. 날짜를 기준으로 정렬 (최신 리뷰를 남기고 싶으면 ascending=False)
df = df.sort_values(by=['작성자', '날짜'])

indices_to_drop = set()

# 2. 작성자별로 그룹화하여 비교
for author, group in df.groupby('작성자'):
    idx_list = group.index.tolist()
    if len(idx_list) < 2: continue  # 리뷰가 하나면 통과
    
    # 그룹 내에서 두 개씩 짝지어 비교
    for i, j in combinations(idx_list, 2):
        sim = get_jaccard_sim(df.loc[i, '리뷰내용_정제'], df.loc[j, '리뷰내용_정제'])
        
        # 유사도가 0.8(80%) 이상이면 중복으로 판단
        if sim >= 0.8:
            # i가 이전 날짜이므로 j(나중 날짜)를 삭제 (원하는 기준에 따라 선택)
            indices_to_drop.add(j)

# 3. 최종 데이터 생성
df_final = df.drop(index=list(indices_to_drop))

print(f"기존 데이터: {len(df)}개")
print(f"유사 중복 제거 후: {len(df_final)}개")
```
